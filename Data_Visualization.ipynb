```python
import matplotlib.pyplot as plt


months = ['Jan', 'Feb', 'Mar', 'Apr']
sales = [100, 200, 150, 300]

plt.bar(months, sales)
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [40, 30, 20, 10]
labels = ['A', 'B', 'C', 'D']

plt.pie(data, labels=labels, autopct='%1.1f%%')
plt.title("Distribution")
plt.show()
```
