```python
sns.boxplot(x='day', y='total_bill', data=df)
plt.show()
```

```python
- Dataset has no missing values
- Most total_bill values are between 10 and 20
- There is a positive relationship between total_bill and tip
- Male customers are more than female customers
- Weekend days have higher total bills
```
